```python
from langchain.text_splitter import CharacterTextSplitter
from langchain.document_loaders import TextLoader

# with open("file1.txt") as f:
#     text = f.read()

loader = TextLoader("file1.txt")
documents = loader.load()
   
text_splitter = CharacterTextSplitter(chunk_size=5, chunk_overlap=0)
text = text_splitter.split_documents(documents)
text
```

```python
deployment_name = "dpmnt001"
openai.api_type = "azure"
openai.api_key = config["OPENAI_API_KEY"]
openai.api_base = config["ENDPOINT"]
openai.api_version = "2023-05-15"
```

```python
import chromadb

from langchain.embeddings.openai import OpenAIEmbeddings
from langchain.llms import AzureOpenAI
from langchain.chains import RetrievalQA
from langchain.vectorstores import Chroma


```

```python
embeddings = OpenAIEmbeddings(deployment=deployment_name,
                  model="gpt-35-turbo",
                  openai_api_version = '2023-05-15',
                  openai_api_base= config["ENDPOINT"],
                  openai_api_key=config["OPENAI_API_KEY"],
                  openai_api_type="azure"
                  )

vectordb = Chroma(persist_directory="./chroma_db", embedding_function=embeddings)
vectorstore = Chroma.from_documents(text, embeddings)

llm = AzureOpenAI(deployment_name=deployment_name,

                  model_name="gpt-35-turbo",
                  openai_api_version = '2023-05-15'
                  
                )

chain = RetrievalQA.from_chain_type(llm=llm,
                                    retriever=vectorstore.as_retriever(),
                                    chain_type="stuff"
)
```

```python
query = "Why was the math book sad"
chain.run(query)
```
